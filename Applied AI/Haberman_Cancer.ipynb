## PDF

```python
alive = pd.DataFrame([hb.iloc[i,:] for i in range(hb.shape[0]) if hb["survival"][i]==1 ])
dead = pd.DataFrame([hb.iloc[i,:] for i in range(hb.shape[0]) if hb["survival"][i]==2 ])
```

```python
count,bin_edges = np.histogram(alive['age'],bins=10)
pdf = count / sum(count)
cdf = np.cumsum(count)
```

```python
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.show()
```
