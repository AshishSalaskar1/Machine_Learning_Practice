## CDF

```python
count,bin_edges = np.histogram(iris_setosa['petal_length'],bins=10)
```

```python
print(count)
print(bin_edges)
```

```python
plt.plot(bin_edges[1:],count)
```

```python
pdf = count / (sum(count))
print(pdf)
```

```python
print("Documents")
```

```python
# Cumulative Sum
cdf = np.cumsum(pdf)
print(cdf)
```

```python
## PDF vs v 
```

```python
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.show()
```
