```python
import numpy as np
import matplotlib.pyplot as plt
import pylab
import scipy.stats as stats
```

## QQ-Plots are used to check if given data follows GAUSSIAN DISTRIBUTION

```python
## Reference data which is N(0,1) -> Y

#normal(mean,standard_deviation,size)
std_norm = np.random.normal(loc=0,scale=1,size=1000)
# for i in range(1,101):
#     print("%d percentile :%f "%(i, np.percentile(std_norm,i)) )
```

#### stats.probplot automatically select Y of size(n) and plots QQ-PLot

```python
#TEST DATA WHICH IS GAUSSIAN N(20,5)
test_data = np.random.normal(loc = 20,scale = 5,size=100)
stats.probplot(test_data, dist="norm", plot=pylab)
pylab.show()
```

##### As points follow a straight line :X (test_data) is similar to Y and
##### X Follows Gaussian Distribution

```python
#TEST DATA WHICH IS UNIFORM N(20,5)
test_data1 = np.random.uniform(low=-1,high=1,size=100)
stats.probplot(test_data1, dist="norm", plot=pylab)
pylab.show()
```

##### As points dont follow a straight line: X(test_data1) is not similar to Y and
##### X doesnt follows Gaussian Distribution
