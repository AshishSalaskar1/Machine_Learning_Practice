```python
import numpy as np
import matplotlib.pyplot as plt
from random import randint
import pylab
import scipy.stats as stats
import math
import seaborn as sns
```

```python
population = np.random.uniform(0,100,3000)
# print(np.mean(population))
```

```python
print("Mean: ",np.mean(population))
print("SD :",np.std(population))
```

```python
# QQ-Plot Test
stats.probplot(population, dist="norm", plot=pylab)
pylab.show()
```

```python
sns.distplot(population)
```

## Create Distribution of means of samples
#### n : points in each sample (>30)
#### n_samples : No of samples taken

```python
n = 50
n_samples = 1000
```

```python
#sample Distr : distribution of means of samples
sample_distr = []
for i in range(n_samples):
    sample = []
    for _ in range(n):
        sample.append(population[randint(0,len(population)-1)])
    sample_distr.append(np.mean(sample))
```

## Expected Results acc to Central Limit Theorem
#### M (Population Mean) , SD : (Population SD)
#### M'= M and SD' = SD / sqrt(n)

```python
print("Mean: ",np.mean(sample_distr))
print("SD :",np.std(sample_distr))
```

## QQ-Plot Test

```python
stats.probplot(sample_distr, dist="norm", plot=pylab)
pylab.show()
```

```python
sns.distplot(sample_distr)
```
